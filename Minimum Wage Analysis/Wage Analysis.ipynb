Notes about the data: 

1. Table_data: The scraped, unclean data from the US Department of Labor. (represents wage/hour in USD)

2. Footnote: Is there some footnote associated with the data?

3. High Value: In instances where Table_data had multiple values, this is the higher of the two values.

4. Low Value: Lower of two values described in high value.

5. CPI Average: This is the average CPI score associated with that year.

6. High 2018: This is the High Value denoted in 2018 adjusted dollars. 

7. Low 2018: This is the Low Value denoted in 2018 adjusted dollars.

Note: Every state and territory has an entry in each year of the dataset. The dataset spans from 1968 to 2017.

```python
calivingwagedf2010 = pd.read_csv('caliving-wage.csv')
```

```python
calivingwagedf2010['reportyear'].min()
#data appears to only cover the year 2010.
```

```python
calivingwagedf2010 = calivingwagedf2010.loc[calivingwagedf2010['family_type'] == 'Adult2_Child1', ['county_name', 'livingwage']]
```

```python
calivingwagedf2010 = calivingwagedf2010.drop_duplicates(subset = 'county_name')
```

```python
calivingwagedf2010 = calivingwagedf2010.dropna(axis = 0)
```

```python
calivingwagedf2010
#This dataframe represents the county-level living wage in 2010 dollars for all CA counties.
```

```python
calivingwagedf2018 = pd.read_csv('calivingwage2018.csv')
```

```python
calivingwagedf2018 = calivingwagedf2018.dropna()
```

```python
calivingwagedf2018
#This dataframe represents the county-level living wage in 2018 dollars for all CA counties.
```

```python
cal2010 = minwagedf[minwagedf['Year'] == 2010]
```

The California minimum wage in 2010 was $\$8.00$ per hour. In 2018 dollars, this is $\$9.19$ per hour. 

The California minimum wage in 2018 was $\$11.00$ per hour. 

```python
statutorywage2010 = [8 for x in range(len(calivingwagedf2010['county_name']))]
statutorywage2018 = [11 for y in range(len(calivingwagedf2018['County Name']))]
```

```python
calivingwagedf2010['State Minimum Wage'] = statutorywage2010
calivingwagedf2018['State Minimum Wage'] = statutorywage2018
```

```python
calivingwagedf2010['Percent Difference (%)'] = ((calivingwagedf2010['livingwage'] - calivingwagedf2010['State Minimum Wage'])/calivingwagedf2010['State Minimum Wage']) * (100)
```

```python
calivingwagedf2010
```

```python
calivingwagedf2018['State Minimum Wage'] = calivingwagedf2018['State Minimum Wage'].astype(float)
```

```python
calivingwagedf2018['State Minimum Wage'].dtype
```

```python
calivingwagedf2018['Wage per hour'].dtype
```

```python
diff = ((calivingwagedf2018['Wage per hour'] - calivingwagedf2018['State Minimum Wage'])/calivingwagedf2018['State Minimum Wage']) * 100
```

```python
calivingwagedf2018['Percent Difference (%)'] = diff
```

```python
calivingwagedf2018
```

```python
caextension = pd.read_csv('extension of the california county datasets.csv')
```

```python
caextension
# I collected additional data on county-level poverty rates, and created an indicator variable for whether 
# the federal government classifies the county as urban or rural (1 = urban, 0 = rural).
```

```python
cafeatures = [[caextension['2016 Poverty Rate (%)'][i], caextension['Urban/Rural Indicator Variable'][i]] for i in range(len(caextension['2016 Poverty Rate (%)']))]
```

```python
y = calivingwagedf2018['Percent Difference (%)']
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg = linreg.fit(cafeatures, y)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

Predicted Model: Percent_Difference($\%$) = $-3.72723883$ $*$ 2016_poverty_rate($\%$) + $27.96158136$ $*$ Urban/Rural_Indicator_Variable + $164.78775148021242$

Logically, this result makes sense. A rural county with a high poverty rate should have a low percent difference, whereas a low-poverty urban county should have a high percent difference. This regression result matches my original intuition---the booming urban areas of California are expensive to live in, and there is a higher difference between the state minimum wage and the estimated hourly wage a worker would have to make for a family to live an bare subsistence life above the poverty line.

```python
linreg.score(cafeatures, y)
```

```python
linreg.predict(cafeatures)
```

```python
from sklearn.feature_selection import f_regression
freg = f_regression(X = cafeatures, y = y)
```
