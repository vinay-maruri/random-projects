```python
import seaborn as sns
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
from pathlib import Path


# Default plot configurations
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,8)
plt.rcParams['figure.dpi'] = 150
sns.set()

from IPython.display import display, Latex, Markdown

excel_file = 'nba1718.csv'
nba = pd.read_csv(excel_file)
```

```python
nba.rename(index = str, columns = {'Rk': 'number in dataset'})
```

```python
nba.sort_values(by = 'MP', ascending = False)
```

```python
nba.sort_values(by = 'AST', ascending = False)
```

```python
nba.sort_values(by = 'BLK', ascending = False)
```

```python
nba.sort_values(by = 'PS/G', ascending = False)
```

```python
nba.sort_values(by = 'FGA', ascending = False)
```

```python
nba['vinayoffscore'] = nba['GS'] * nba['MP'] * nba['FGA'] * nba['FG'] * nba['ORB'] * nba['PS/G'] * nba['FG'] * nba['3P'] * nba['FT'] * nba['AST']
```

```python
nba.sort_values(by = 'vinayoffscore', ascending = False)
```

```python
nba['vinaydefscore'] = nba['STL'] * nba['BLK'] * nba['DRB'] * nba ['PF']
```

```python
nba.sort_values(by = 'vinaydefscore', ascending = False)
```

```python
rankings = nba.loc[:, ['Player', 'Pos', 'Age', 'vinayoffscore', 'vinaydefscore']]
```

```python
rankings['vinayoverallscore'] = rankings['vinayoffscore'] + rankings['vinaydefscore']
```

```python
rankings.sort_values(by = 'vinayoverallscore', ascending = False)
```

```python
finalrankings = rankings.loc[:, ['Player', 'Pos', 'Age', 'vinayoverallscore']]
```

```python
finalrankings['rank'] = finalrankings['vinayoverallscore'].rank()
```

```python
finalrankings.sort_values(by = 'rank', ascending = False)
```

```python
writer = pd.ExcelWriter('finalrankings.xlsx')
finalrankings.to_excel(writer, 'Sheet1')
nba.to_excel(writer, 'Sheet2')
```

```python
writer.save()
```
